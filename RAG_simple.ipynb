```python
# https://zhuanlan.zhihu.com/p/668082024
# 从txt文本中加载数据
from langchain.document_loaders import TextLoader
loader = TextLoader("./藜麦 - 副本.txt")
documents = loader.load()
```

```python
# 数据的形态
print(documents)
```

```python
# 文档分割
from langchain.text_splitter import CharacterTextSplitter


# 创建拆分器
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50,separator="\n", length_function=len)
# 用拆分器拆分文档
documents = text_splitter.split_documents(documents)
```

```python
# 看一下拆分之后的文本
for i in documents:
    print(i)
    print('\n')
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

# 加载向量化模型（用于将上面的每一段文本和后面用户的问题转换为向量，用于在检索时匹配用户问题）
# embedding model: m3e-base
model_name = "E:\LLM\model\Jerry0\m3e-base"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}
embedding = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs,
)
```

```python
# load data to Chroma db
# Chroma是一个轻量级的向量数据库
from langchain.vectorstores import Chroma
# 将上面切分后的文本通过前面说到的向量化模型存到向量库里面
db = Chroma.from_documents(documents, embedding)
```

```python
# similarity search
# 用向量库进行相似性搜索，即把用户问题和向量库里面存储内容相关性高的找出来
# 女职工多少天产检？

similarity_search_result = db.similarity_search("工作第2年了，我的探亲假如何计算？")
ss_result = [x.page_content for x in similarity_search_result]
ss_string = ''
num = 0
for i in ss_result:
    num = num + 1
    ss_string  = "结果" + str(num) + '\n' + i + '\n'
    print(ss_string)
```
